```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
```

```python
data["mfr"].unique
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['calories'], bins=10, edgecolor='black')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Histogram of Calories')
```

### Most common calorie range in the cereal dataset is between approximately 100 and 150 calories

```python
# Styling Artists
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data['rating'], data['calories'], color='brown', linewidth=2, linestyle='--', marker='o', markersize=6)
ax.set_title('Calories vs Rating')
plt.text(60, 130, 'Clustered', fontsize=15, color='black')
plt.show()
```

### There is no clear trend between calories and rating.

![](https://media.tenor.com/ToG4_NJCnf4AAAAC/ugh-ughh.gif)

```python
top_cereals = data.sort_values('rating', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_cereals['name'], top_cereals['rating'], color='green')
plt.xlabel('Ratings')
plt.title('Top 10 Cereals by Rating')
```

### The cereals in the top 10 have a small range of quality among as most ratings are below 80

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(data['type'].value_counts().index,data['type'].value_counts().values,color=['#fb9a99', '#e31a1c'])
ax.set_title('Cereal Type')
```

### There are more cold cereals than hot cereals in the dataset

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data['calories'],data['sugars'],color='black')
ax.set_xlabel('Calories')
ax.set_ylabel('Sugar Content')
ax.set_title('Sugar vs. Calories')
plt.show()
```

 ### There is a positive correlation between the calories and sugar content in cereals.However, there are some cereals with low calories but high sugar content, and vice versa.

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(data["mfr"], data["shelf"], color=['#a6cee3', '#1f78b4','#b2df8a'], label=['Shelf Life'])
ax.set(xlabel='Manufacturer' , ylabel = 'Shelf Life (months)',title='Manufacturer vs Shelf Life')
ax.legend(loc="upper left")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(data['protein'], data['sodium'], color='#f28e2b')
ax.set_xlabel('Protein')
ax.set_ylabel('Sodium')
ax.set_title('Protein vs. Sodium in Cereal')
plt.show()
```

### There is no significant correlation between protein and sodium content in cereals.

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(data['protein'], color='#e15759')
ax.set_xlabel('Protein Content')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Protein Content')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['mfr'], data['fiber'], label='Fiber')
plt.bar(data['mfr'], data['sugars'], bottom=data['fiber'], label='Sugars')
plt.xticks(rotation=75)
plt.legend()
plt.title('Fiber and Sugars by Manufacturer')
```

### Manufacture N has more fiber content

```python
plt.figure(figsize=(12, 6))
plt.bar(data['name'], data['sugars'], label='Sugars', color='purple')
plt.bar(data['name'], data['fiber'], bottom=data['sugars'], label='Fiber', color='#4e79a7')
plt.xlabel('Cereal')
plt.ylabel('Grams')
plt.title('Sugar and Fiber Content in Cereal')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

### The cereal with the highest amount of sugar is Froot Loops, while the cereal with the highest amount of fiber is All-Bran.On the other hand, some cereals are high in sugar but low in fiber, such as Froot Loops and Golden Crisp.

![](https://media.tenor.com/Ae4v0kOKmxEAAAAC/thats-it-pretty-impressive-bgt.gif)

```python
fig, axd = plt.subplot_mosaic([['calories', 'sugars'], ['fat', 'carbo']], figsize=(10,6), layout='constrained')

axd['calories'].boxplot(data['calories'])
axd['calories'].set_title('Calories')

axd['sugars'].boxplot(data['sugars'])
axd['sugars'].set_title('Sugars')

axd['fat'].boxplot(data['fat'])
axd['fat'].set_title('Fat')

axd['carbo'].boxplot(data['carbo'])
axd['carbo'].set_title('Carbohydrates')

plt.suptitle('Boxplots of Nutrient Content')
plt.tight_layout()
plt.show()
```

### The median calorie content is around 110, with some cereals having very high calorie counts. The sugar content is also varied, with some cereals having very high amounts of sugar. The fat content is lower than the calorie content.

```python
# Select data for line plot
data_for_plot = data.loc[data['mfr'].isin(['K', 'N'])].groupby('mfr').mean()[['calories', 'sugars','fat','carbo']]

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(12, 6))

# Create lines with different styles and widths
ax.plot(data_for_plot.index, data_for_plot['calories'], linestyle='-', linewidth=2, label='Calories')
ax.plot(data_for_plot.index, data_for_plot['sugars'], linestyle='--', linewidth=3, label='Sugars')
ax.plot(data_for_plot.index, data_for_plot['fat'], linestyle='--', linewidth=4, label='Fat')
ax.plot(data_for_plot.index, data_for_plot['carbo'], linestyle='-', linewidth=5, label='Carbohydrates')

ax.set_xlabel('Manufacturer')
ax.set_ylabel('Content')
ax.set_title('Calories and Sugars in Cereal')
ax.legend()
plt.show()
```

### Cereal from manufacturer N has a higher content of calories, sugars, and fat compared to cereal from manufacturer K. On the other hand, cereal from manufacturer K has a higher content of carbohydrates compared to cereal from manufacturer N.

![](https://media.tenor.com/IM5h4sa5ofYAAAAd/ellen-pompeo-keep-going.gif)

```python
nutrient_df = pd.pivot_table(data, values=['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins'], index='mfr')
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
axs = axs.flatten()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']
for i, col in enumerate(nutrient_df.columns):
    axs[i].bar(nutrient_df.index, nutrient_df[col],color=colors[i])
    axs[i].set_xlabel('Manufacturer')
    axs[i].set_ylabel(col.capitalize())
    axs[i].set_title(f'{col.capitalize()} Content')
    axs[i].tick_params(axis='x', labelrotation=90)
plt.tight_layout()
plt.show()
```

### Kellogg's (K) and General Mills (G) have the highest nutrient content across most categories, while Quaker Oats (Q) and Nabisco (N) have the lowest.
### Quaker Oats (Q) has the highest fiber content, while Nabisco (N) has the highest carbohydrate and sugar content.The manufacturers have relatively similar sodium and fat content.

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Plot shelf life data with red dashes
ax.plot(data['name'], data['shelf'], 'r--', label='Shelf Life')

# Add blue squares for ratings data
ax.plot(data['name'], data['rating'], 'bs', label='Ratings')

# Add green triangles for sugar data
ax.plot(data['name'], data['sugars'], 'g^', label='Sugars')

ax.set_xlabel('Cereal Name')
ax.set_ylabel('Value')
ax.set_title('Cereal Data with Different Markers')
plt.text(10, 75, 'Impressive Chart', fontsize=15, color='orange')
ax.legend()
plt.xticks(rotation=90)
plt.show()
```

###  Some cereals have long shelf lives while others have short shelf lives, regardless of their ratings or sugar content.

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data['mfr'], data['sodium'],color='#76b7b2')
ax.set_xlabel('Manufacturer')
ax.set_ylabel('Sodium Content')
ax.set_title('Sodium Content over Time')
plt.xticks(rotation=90)
plt.show()
```

### The sodium content is relatively consistent over time for each manufacturer, with some small fluctuations while manufacturer "G" have highest sodium content and manufacturer "K" have lowest sodium content.

```python
names = data["type"]
values = data["shelf"]
plt.figure(figsize=(10, 5))

# Our First Plot
plt.subplot(131)
plt.bar(names, values,color='#ff9da7')
plt.title('Bar Plot')
plt.xlabel('Cereal Type')
plt.ylabel('Shelf Life')

# Our Second Plot
plt.subplot(132)
plt.scatter(names, values,color='#59a14f')
plt.title('Scatter Plot')
plt.xlabel('Cereal Type')

# Our Third Plot
plt.subplot(133)
plt.plot(names, values,color='#b07aa1')
plt.title('Line Plot')
plt.xlabel('Cereal Type')
plt.suptitle('Categorical Plotting')

plt.show()
```

### Different types of cereal have varying shelf life values. The scatter plot shows more detail and variation in the data points than the bar chart or line plot.

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot of nutrient content
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['calories'], data['fiber'], data['sugars'])
ax.set_xlabel('Calories')
ax.set_ylabel('Fiber Content')
ax.set_zlabel('Sugar Content')
ax.set_title('Relationship between Calories, Fiber, and Sugar')
ax.legend(['Cereal'], loc='upper left')
plt.show()
```

### Most cereals are clustered around the lower end of the calorie axis, with a few outliers with higher calorie content.There is a general trend towards higher fiber content as calorie content decreases.Sugar content varies widely across cereals, with some having very low sugar content and others having much higher levels.

## Observations

### Sugar content: Many cereals have a high sugar content, with some brands containing more than 20g of sugar per serving. It is important to check the label and choose cereals with lower sugar content to maintain a healthy diet.

### Caloric content: The calorie content of cereals varies widely, with some containing more than 120 calories per serving, while others contain less than 50. It is important to choose cereals with lower calorie content to maintain a healthy weight.

### Manufacturer: There are several manufacturers of breakfast cereals, with some brands being more popular than others. Kellogg's is the largest manufacturer, followed by General Mills and Post.

### Nutrient content: The nutrient content of cereals can vary widely, with some brands containing high amounts of vitamins and minerals, while others may be lacking in certain nutrients. It is important to choose cereals with a balanced nutrient profile to ensure that dietary requirements are met.

### Shelf life: The shelf life of cereals varies depending on the brand, with some lasting for several months, while others only last a few weeks. It is important to check the expiration date and consume cereals before they go bad.
